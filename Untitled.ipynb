```python
import numpy as np
import pandas as pd
import os
```

```python
def get_data(file, dt, gc_or_not = False) :
    n_data = np.load(os.path.join(file, f'data_{dt}.npy'))
    f_data_raw = np.load(os.path.join(file, f'features_{dt}.npy'), allow_pickle=True)

    f_data_dict = {'Played_frequency':[], 'Condition':[], 'Block':[], 'Frequency_changes':[], 'Mock_frequency':[], 'Mock_change':[]}
    for item in f_data_raw:
        for key, value in item.items():
            f_data_dict[key].append(value)
    f_data = pd.DataFrame(f_data_dict)
    return n_data,f_data
```

```python

directory = "C:/Users/PenPen/Desktop/Ferret/Data/"
dt = 0.02

n_data,f_data = get_data(directory, dt, gc_or_not = False)
```
